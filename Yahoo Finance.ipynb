```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,6

# Set stock and data source
stock = 'VOW3.DE'
source = 'yahoo'

# Set date range (Google went public August 19, 2004)
start = datetime.datetime(2004, 8, 19)
end = datetime.datetime(2017, 4, 24)

# Collect Google stock data
goog_df = data.DataReader(stock, source, start, end)
goog_df.head()
```

```python
# Plot GOOG adjusted close for 2016
goog_df['Adj Close'].plot(kind='line', grid=True, title='GOOG Adjusted Closes, IPO through 2016')
```

```python
from yahoo_finance import Share
```

```python
share = Share(stock)
print(share.get_name())
print('P/B: ' , share.get_price_book())
print('P/E: ' , share.get_price_earnings_ratio())
print('EBITDA: ' , share.get_ebitda())

print()

print('Price: ' , share.get_price())
print('200-Day Mvg Avg: ' , share.get_200day_moving_avg())
print('Pct Change from 200-Day Mvg Avg: ' , share.get_percent_change_from_200_day_moving_average())

print()
print('Dividend Yield: ', share.get_dividend_yield())
print('Dividend Share: ' , share.get_dividend_share())
print('Dividend Paydate: ' , share.get_dividend_pay_date())
print('Short Ratio: ' , share.get_short_ratio())

print()
print('Market Cap: ' , share.get_market_cap())
print('Volume: ' , share.get_volume())
print('Daily Avg Volume: ' , share.get_avg_daily_volume())
```
